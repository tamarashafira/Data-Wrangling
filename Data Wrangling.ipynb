## HANDLE MISSING DATA     

Terdapat beberapa isian yang menandakan missing value, seperti: n/a, NA, -, null, kosong. 

```python
# Read data with missing value

missing_values = ["na", "NA", "-", "NULL"]
hrdata = pd.read_excel("HRDataset.xlsx", na_values = missing_values)
```

```python
# checking null value

hrdata.isnull().any()
```

```python
# checking NA value
# jika terdapat false maka pada kolom tersebut tidak ada data yang kosong

hrdata.isna().any()
```

```python
# total missing value

hrdata.isna().sum()
```

```python
# Copy data frame

hrdata1 = hrdata.copy()
```

```python
hrdata1["Pay Rate"]
```

```python
hrdata[hrdata['Date of Termination'].isnull()]
```

### Hadling Missing Value
